## Advanced Options

Beyond local editing two things are quite important: editing the notebooks in the markdown format and running Jupyter remotely.
The latter matters when we want to run the code on a faster server.
The former matters since Jupyter's native ipynb format stores a lot of auxiliary data that is
irrelevant to the content,
mostly related to how and where the code is run.
This is confusing for Git, making
reviewing contributions very difficult.
Fortunately there is an alternative---native editing in the markdown format.

### Markdown Files in Jupyter

If you wish to contribute to the content of this book, you need to modify the
source file (md file, not ipynb file) on GitHub.
Using the notedown plugin we
can modify notebooks in the md format directly in Jupyter.


First, install the notedown plugin, run the Jupyter Notebook, and load the plugin:

```
pip install d2l-notedown  # You may need to uninstall the original notedown.
jupyter notebook --NotebookApp.contents_manager_class='notedown.NotedownContentsManager'
```

You may also turn on the notedown plugin by default whenever you run the Jupyter Notebook.
First, generate a Jupyter Notebook configuration file (if it has already been generated, you can skip this step).

```
jupyter notebook --generate-config
```

Then, add the following line to the end of the Jupyter Notebook configuration file (for Linux or macOS, usually in the path `~/.jupyter/jupyter_notebook_config.py`):

```
c.NotebookApp.contents_manager_class = 'notedown.NotedownContentsManager'
```

After that, you only need to run the `jupyter notebook` command to turn on the notedown plugin by default.

### Running Jupyter Notebooks on a Remote Server

Sometimes, you may want to run Jupyter notebooks on a remote server and access it through a browser on your local computer. If Linux or macOS is installed on your local machine (Windows can also support this function through third-party software such as PuTTY), you can use port forwarding:

```
ssh myserver -L 8888:localhost:8888
```

The above string `myserver` is the address of the remote server.
Then we can use http://localhost:8888 to access the remote server `myserver` that runs Jupyter notebooks. We will detail on how to run Jupyter notebooks on AWS instances
later in this appendix.

### Timing

We can use the `ExecuteTime` plugin to time the execution of each code cell in Jupyter notebooks.
Use the following commands to install the plugin:

```
pip install jupyter_contrib_nbextensions
jupyter contrib nbextension install --user
jupyter nbextension enable execute_time/ExecuteTime
```

````python
import matplotlib.pyplot as plt
from PIL import Image

# Fungsi untuk membaca gambar JPG dan mengembalikan data pikselnya
def read_jpg(filename):
    # Buka gambar JPG menggunakan PIL
    img = Image.open(filename)
    # Konversi gambar ke format RGB jika perlu
    img = img.convert('RGB')
    # Ambil dimensi gambar (lebar dan tinggi)
    width, height = img.size
    # Ambil data piksel dalam bentuk list
    data = list(img.getdata())
    return width, height, data

# Fungsi untuk menulis data piksel ke gambar dan menyimpannya sebagai JPG
def write_jpg(filename, width, height, data):
    # Buat gambar baru dengan dimensi yang diberikan
    img = Image.new('RGB', (width, height))
    # Masukkan data piksel yang telah diproses
    img.putdata(data)
    # Simpan gambar sebagai file JPG
    img.save(filename, 'JPEG')

# Fungsi untuk menyesuaikan kecerahan gambar
def adjust_brightness(data, factor):
    return [(min(int(pixel[0] * factor), 255),
             min(int(pixel[1] * factor), 255),
             min(int(pixel[2] * factor), 255)) for pixel in data]

# Fungsi untuk menyesuaikan kontras gambar
def adjust_contrast(data, factor):
    # Menghitung rata-rata untuk setiap channel (merah, hijau, biru)
    mean = tuple(sum(channel) // len(channel) for channel in zip(*data))
    return [(min(max(int((pixel[0] - mean[0]) * factor + mean[0]), 0), 255),
             min(max(int((pixel[1] - mean[1]) * factor + mean[1]), 0), 255),
             min(max(int((pixel[2] - mean[2]) * factor + mean[2]), 0), 255)) for pixel in data]

# Fungsi untuk menyesuaikan intensitas warna gambar
def adjust_color(data, factor):
    return [(min(int(pixel[0] * factor), 255),
             min(int(pixel[1] * factor), 255),
             min(int(pixel[2] * factor), 255)) for pixel in data]

# Fungsi untuk mengasah gambar menggunakan filter sederhana
def sharpen(data, width, height):
    # Fungsi untuk mengambil nilai piksel dengan memastikan berada dalam batas gambar
    def get_pixel(x, y):
        if x < 0 or x >= width or y < 0 or y >= height:
            return (0, 0, 0)  # Nilai default untuk piksel di luar batas
        idx = y * width + x
        return data[idx]

    # Salin data piksel untuk diubah
    new_data = data[:]
    # Iterasi melalui setiap piksel dalam gambar
    for y in range(height):
        for x in range(width):
            idx = y * width + x
            center = get_pixel(x, y)
            # Ambil piksel tetangga di sekitarnya
            neighbors = (
                get_pixel(x-1, y), get_pixel(x+1, y),
                get_pixel(x, y-1), get_pixel(x, y+1)
            )
            # Terapkan filter sharpen (pengasahan)
            sharpened = (
                min(max(2 * center[0] - sum([n[0] for n in neighbors]) // 4, 0), 255),
                min(max(2 * center[1] - sum([n[1] for n in neighbors]) // 4, 0), 255),
                min(max(2 * center[2] - sum([n[2] for n in neighbors]) // 4, 0), 255)
            )
            new_data[idx] = sharpened
    return new_data

# Main program
# Baca gambar awal (gunakan JPG sebagai input)
input_image_path = "gedung.jpg"
width, height, data = read_jpg(input_image_path)

# Proses gambar dengan menyesuaikan kecerahan, kontras, warna, dan pengasahan
bright_data = adjust_brightness(data, 1.3)  # Penyesuaian kecerahan
contrast_data = adjust_contrast(bright_data, 1.1)  # Penyesuaian kontras
color_data = adjust_color(contrast_data, 1.4)  # Penyesuaian warna
sharpened_data = sharpen(color_data, width, height)  # Pengasahan gambar

# Tampilkan gambar sebelum dan sesudah pemrosesan menggunakan matplotlib
input_img = Image.open(input_image_path)  # Gambar input
output_img = Image.new('RGB', (width, height))  # Gambar output

# Masukkan data piksel yang telah diproses ke gambar output
output_img.putdata(sharpened_data)

# Menampilkan gambar sebelum dan sesudah pemrosesan berdampingan
plt.figure(figsize=(12, 6))

# Gambar asli
plt.subplot(1, 2, 1)
plt.title("Gambar sebelum")
plt.axis("off")
plt.imshow(input_img)

# Gambar setelah pemrosesan
plt.subplot(1, 2, 2)
plt.title("Gambar sesudah")
plt.axis("off")
plt.imshow(output_img)

# Menampilkan plot
plt.show()

# Cetak pesan bahwa proses selesai
print("Proses selesai, hasil telah ditampilkan.")
````

````python
#MODUL6TUGAS2

from PIL import Image, ImageDraw, ImageFont
import matplotlib.pyplot as plt

# Fungsi untuk mengatur kecerahan gambar
def atur_kecerahan(gambar, faktor):
    gambar_data = gambar.load()  # Mengambil data piksel gambar
    for y in range(gambar.height):  # Iterasi untuk setiap baris (height)
        for x in range(gambar.width):  # Iterasi untuk setiap kolom (width)
            r, g, b = gambar_data[x, y]  # Mendapatkan nilai RGB dari setiap piksel
            r = min(255, max(0, int(r * faktor)))  # Mengatur kecerahan untuk merah (r)
            g = min(255, max(0, int(g * faktor)))  # Mengatur kecerahan untuk hijau (g)
            b = min(255, max(0, int(b * faktor)))  # Mengatur kecerahan untuk biru (b)
            gambar_data[x, y] = (r, g, b)  # Menyimpan nilai baru untuk piksel
    return gambar  # Mengembalikan gambar yang sudah diubah

# Fungsi untuk mengubah ukuran gambar
def ubah_ukuran(gambar, ukuran):
    return gambar.resize(ukuran)  # Mengubah ukuran gambar sesuai dengan parameter ukuran

# Fungsi untuk menambahkan border pada gambar
def add_border(gambar, ukuran_bingkai, warna):
    lebar, tinggi = gambar.size  # Mendapatkan ukuran gambar (lebar dan tinggi)
    gambar_baru = Image.new("RGB", (lebar + 2 * ukuran_bingkai, tinggi + 2 * ukuran_bingkai), warna)  # Membuat gambar baru dengan border
    gambar_baru.paste(gambar, (ukuran_bingkai, ukuran_bingkai))  # Menempelkan gambar asli ke tengah gambar baru
    return gambar_baru  # Mengembalikan gambar dengan border

# Fungsi untuk tajamkan gambar
def tajamkan_gambar(gambar):
    lebar, tinggi = gambar.size  # Mendapatkan ukuran gambar
    gambar_data = gambar.load()  # Mendapatkan data piksel gambar
    gambar_tajam = gambar.copy()  # Membuat salinan gambar
    gambar_tajam_data = gambar_tajam.load()  # Mendapatkan data piksel salinan gambar

    # Iterasi untuk setiap piksel (kecuali pinggiran)
    for y in range(1, tinggi - 1):
        for x in range(1, lebar - 1):
            # Menghitung nilai tajam dengan metode filter sederhana (seperti filter konvolusi)
            r = 5 * gambar_data[x, y][0] - (gambar_data[x-1, y][0] + gambar_data[x+1, y][0] +
                                             gambar_data[x, y-1][0] + gambar_data[x, y+1][0])
            g = 5 * gambar_data[x, y][1] - (gambar_data[x-1, y][1] + gambar_data[x+1, y][1] +
                                             gambar_data[x, y-1][1] + gambar_data[x, y+1][1])
            b = 5 * gambar_data[x, y][2] - (gambar_data[x-1, y][2] + gambar_data[x+1, y][2] +
                                             gambar_data[x, y-1][2] + gambar_data[x, y+1][2])
            # Membatasi nilai RGB agar tetap berada dalam rentang 0-255
            r = min(255, max(0, r))
            g = min(255, max(0, g))
            b = min(255, max(0, b))
            gambar_tajam_data[x, y] = (r, g, b)  # Menyimpan nilai tajam ke salinan gambar

    return gambar_tajam  # Mengembalikan gambar yang sudah tajam

# Fungsi untuk mengubah gambar menjadi grayscale
def ubah_ke_grayscale(gambar):
    lebar, tinggi = gambar.size  # Mendapatkan ukuran gambar
    gambar_data = gambar.load()  # Mendapatkan data piksel gambar
    for y in range(tinggi):  # Iterasi untuk setiap baris
        for x in range(lebar):  # Iterasi untuk setiap kolom
            r, g, b = gambar_data[x, y]  # Mendapatkan nilai RGB
            grayscale = int((r + g + b) / 3)  # Menghitung nilai grayscale dengan rata-rata RGB
            gambar_data[x, y] = (grayscale, grayscale, grayscale)  # Mengganti nilai RGB dengan grayscale
    return gambar  # Mengembalikan gambar yang sudah dalam grayscale

# Fungsi untuk menambahkan watermark pada gambar
def tambahkan_watermark(gambar, teks_watermark):
    draw = ImageDraw.Draw(gambar)  # Membuat objek untuk menggambar pada gambar
    font = ImageFont.load_default()  # Menggunakan font default

    # Mencoba untuk memuat font TrueType dengan ukuran yang lebih besar
    try:
        font = ImageFont.truetype("arial.ttf", 40)  # Menggunakan font Arial dengan ukuran 40
    except IOError:
        font = ImageFont.load_default()  # Jika font tidak ditemukan, gunakan font default

    lebar, tinggi = gambar.size  # Mendapatkan ukuran gambar

    # Menggunakan textbbox untuk menghitung ukuran teks watermark
    text_bbox = draw.textbbox((0, 0), teks_watermark, font=font)
    text_width = text_bbox[2] - text_bbox[0]  # Menghitung lebar teks
    text_height = text_bbox[3] - text_bbox[1]  # Menghitung tinggi teks

    # Menentukan posisi watermark di pojok kanan bawah
    posisi = (lebar - text_width - 10, tinggi - text_height - 10)
    draw.text(posisi, teks_watermark, font=font, fill="blue")  # Menambahkan watermark pada gambar
    return gambar  # Mengembalikan gambar yang sudah ada watermark

# Fungsi untuk menyimpan gambar ke file
def simpan_gambar(gambar, path_output):
    gambar.save(path_output)  # Menyimpan gambar ke path_output

# Fungsi untuk memproses gambar (menambahkan efek, watermark, dll.)
def proses_gambar(path_gambar, path_output, teks_watermark, dalam_grayscale=False):
    gambar = Image.open(path_gambar)  # Membuka gambar dari file

    gambar = atur_kecerahan(gambar, 1.3)  # Mengatur kecerahan gambar

    gambar = ubah_ukuran(gambar, (300, 300))  # Mengubah ukuran gambar

    gambar = add_border(gambar, 20, (0, 0, 0))  # Menambahkan border hitam

    gambar = tajamkan_gambar(gambar)  # Menajamkan gambar

    if dalam_grayscale:
        gambar = ubah_ke_grayscale(gambar)  # Mengubah gambar ke grayscale jika diperlukan

    gambar = tambahkan_watermark(gambar, teks_watermark)  # Menambahkan watermark

    simpan_gambar(gambar, path_output)  # Menyimpan gambar hasil ke file

    return gambar  # Mengembalikan gambar yang sudah diproses

# Fungsi untuk menampilkan gambar asli dan hasilnya dalam satu plot
def tampilkan_gambar(path_asli, gambar_hasil):
    gambar_asli = Image.open(path_asli)  # Membuka gambar asli

    # Menampilkan gambar asli dan hasil dalam satu plot
    plt.figure(figsize=(10, 5))

    plt.subplot(1, 2, 1)  # Gambar asli di sebelah kiri
    plt.title("Gambar Asli")
    plt.axis("off")  # Menyembunyikan sumbu
    plt.imshow(gambar_asli)

    plt.subplot(1, 2, 2)  # Gambar hasil di sebelah kanan
    plt.title("Gambar Hasil")
    plt.axis("off")  # Menyembunyikan sumbu
    plt.imshow(gambar_hasil)

    plt.show()  # Menampilkan plot

if __name__ == "__main__":
    gambar_input = "bakso.jpg"  # Gambar input yang akan diproses
    gambar_output = "Hasil.jpg"  # Path untuk menyimpan gambar hasil
    watermark = "BEIMMMMMBBB"  # Teks watermark yang akan ditambahkan

    hasil = proses_gambar(gambar_input, gambar_output, watermark, dalam_grayscale=False)  # Memproses gambar
    tampilkan_gambar(gambar_input, hasil)  # Menampilkan gambar asli dan hasil
````
